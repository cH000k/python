```python
app = JupyterDash(__name__, external_stylesheets=[dbc.themes.MATERIA])

# Create server variable with Flask server object for use with gunicorn
server = app.server

# Create Dataframe
df = pd.read_csv("pokemon.csv")

# Get Total Rows
count = len(df.index)

# Get Total Num of Legendary Pokemon
legendary_count = df.loc[df["Legendary"] == True].shape[0]

# Get Unique Values of one Column and make it to a list
type1_names = df["Type 1"].unique().tolist()

# Get Count of each value and store to a variable
type1_values = []

for type in type1_names:
    value_count = df.loc[df["Type 1"] == type].shape[0]
    type1_values.append(value_count)

df3 = pd.DataFrame({
    "Types": type1_names,
    "": type1_values
})

# Bar Chart
fig = px.bar(df3, x="Types", y="", barmode="group", text_auto=True)
#Disable/Remove Axis label
#fig.update_yaxes(visible=False, showticklabels=True)

# Card
first_card = dbc.Card(
    dbc.CardBody(
        [
            html.H5("Total Number of Pokemons", className="card-title",  style={'font-size': '2rem'}),
            html.P(count,  style={'font-size': '2rem'}),
        ]
    )
)


second_card = dbc.Card(
    dbc.CardBody(
        [
            html.H5("Number of Legendary Pokemons", className="card-title", style={'font-size': '2rem'}),
            html.P(legendary_count,  style={'font-size': '2rem'}),
        ]
    )
)


cards = dbc.Row(
    [
        dbc.Col(first_card, width=4),
        dbc.Col(second_card, width=4),
    ]
)

# Construct App Layout
app.layout = dbc.Container(
    [
        dbc.Row(dbc.Col(html.Div("A single column"))),
        dbc.Row(
            [
                dbc.Col(children=[
                    html.H5("Total Number of Pokemons", className="card-title",  style={'font-size': '2rem'}),
                    html.P(count,  style={'font-size': '2rem'}),
                    ], width=6),
                dbc.Col(html.Div("One of three columns")),
                dbc.Col(html.Div("One of three columns")),
            ]
        ),
    ]
)

app.run_server()
```
